```python
import os
import random
import shutil

# caminhos (ajusta pro teu caso)
origem_img = "dataset/troncos/test/images"
origem_lbl = "dataset/troncos/test/labels"

destino_img = "dataset/troncos_250/images"
destino_lbl = "dataset/troncos_250/labels"


os.makedirs(destino_img, exist_ok=True)
os.makedirs(destino_lbl, exist_ok=True)

# lista imagens
imagens = [f for f in os.listdir(origem_img) if f.endswith((".jpg", ".png"))]

# pega 2000 aleatórias
selecionadas = random.sample(imagens, 250)

for img in selecionadas:
    # copia imagem
    shutil.copy(os.path.join(origem_img, img), destino_img)

    # copia label correspondente
    label = img.rsplit(".", 1)[0] + ".txt"
    caminho_label = os.path.join(origem_lbl, label)

    if os.path.exists(caminho_label):
        shutil.copy(caminho_label, destino_lbl)

print("✅ 2k imagens copiadas com sucesso!")
```
